# General plot.**py**

```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns # improved version of matplotlib.
```

```
x = np.random.random((10,1)) # 10 rows and 1 column

print(x)
```

```
plt.plot(x, '*-') #the y value was automatically assigned by python.
plt.show()
```

```
x = np.linspace(0, 100, 50) # from 0 to 100 have 50 values in equal interval.
y = np.power(x, 2)
plt.plot(x, y, '+-')
plt.show()
```

```
sns.set() # initializing for seaborn.
sns.lineplot(x =x, y=y)
plt.show()
```

# **Plotting Data from Dataset**

load bigger data much faster.manipulations are performed well.


```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```
data = pd.read_csv('/content/machine_learning/Classification/mobile_cleaned.csv')
data.head() # tail show last 5 entries.
```

```
data.describe() # give more info about data. std determins the variation of the data.
```

# Uni-Variant

types of variants(data analysis method):
univariant-> one feature analysis.
bivariant
multivariant

```
sns.countplot(x ='sim_type', data = data)# counts the occurnece of the unique data.#not suitable for long data.
plt.show() # try to have balance data.
```

```
sns.distplot(data["weight"])
plt.show()
```

```
sns.histplot(data=data, x="stand_by_time") # somewhere between countplot and displot.
plt.show()
```

# Bi-Variant

```
ax = sns.scatterplot(x="stand_by_time", y="battery_capacity", data=data)
plt.show()
```

```
#Minimum, 25,50,75, max
ax = sns.boxplot(x="is_liked", y="battery_capacity", data=data)
plt.show()
#Minimum = Q1–1.5 * IQR and Maximum = Q3 + 1.5 * IQR . the small circle are outliners.
```

```
ax = sns.boxplot(x = "expandable_memory", y = "price", data=data)
plt.show()
```

# Multi-Variant

```
ax = sns.violinplot(x='expandable_memory', y='price', data=data,
               hue='is_liked', split=True)
plt.show()
```

```
ax = sns.scatterplot(x = "stand_by_time", y = "battery_capacity", hue="thickness", data=data)
plt.show()
```

```
dataframe = pd.read_csv("/content/machine_learning/Classification/iris.csv")
print(dataframe.head())


ax = sns.scatterplot(x="petal.length", y="petal.width",hue="Variety",size = "sepal.width",data=dataframe)
plt.show()
```

```
ax = sns.swarmplot(x='sepal.length', y='Variety', data=dataframe)
```

```
sns.pairplot(dataframe[['petal.length','petal.width','sepal.length','sepal.width']])
plt.show()
```

```
sns.pairplot(dataframe, hue='Variety')
plt.show()
```
