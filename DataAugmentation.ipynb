Data Agumentation is the process of increaing the number the images when we are given with smaller datasets.We shift,slip, and roate the image which we are having.


```python
#importing necesaary modules

from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import datasets
import matplotlib.pyplot as plt
```

```python
#creating and loading datasets
(train_images,train_labels),(test_images,test_labels) = datasets.cifar10.load_data()
train_images,test_images = train_images/255.0,test_images/255.0
```

```python
#shifting,fliping,rotating the image to increase the size of the datasets

#datagen is an generator which is going to change the image to the specified format.
datagen = ImageDataGenerator(
rotation_range = 40,
width_shift_range = 0.2,
height_shift_range = 0.2,
shear_range = 0.2,
zoom_range = 0.2,
horizontal_flip = True,
fill_mode = 'nearest'
)

#picking an image to change

test_img = train_images[20]
img = image.img_to_array(test_img) # converting the image to numpy arraty (width,height,channels)
img = img.reshape((1,) + img.shape) # 1 is the batch size

i=0

for batch in datagen.flow(img, save_prefix = "test", save_format ="jpeg"):
  plt.figure(i)
  plot = plt.imshow(image.img_to_array(batch[0])) # imshow takes the numpy array of image as input and display an image in the matplotlib figure.since batch size is 1, the image is accessed by batch[0].
  i+=1
  if i > 3:
    break
plt.show()

```
